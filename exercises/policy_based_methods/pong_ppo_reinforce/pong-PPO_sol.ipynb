# Policy

## Exercise 1: Implement your policy
 
Here, we define our policy. The input is the stack of two different frames (which captures the movement), and the output is a number $P_{\rm right}$, the probability of moving left. Note that $P_{\rm left}= 1-P_{\rm right}$

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

# set up a convolutional neural net
# the output is the probability of moving right
# P(left) = 1-P(right)
class Policy(nn.Module):

    def __init__(self):
        super(Policy, self).__init__()
        
    ########
    ## 
    ## Modify your neural network
    ##
    ########
        
        # 80x80 to outputsize x outputsize
        # outputsize = (inputsize - kernel_size + stride)/stride 
        # (round up if not an integer)
        self.conv1 = nn.Conv2d(2, 4, kernel_size=2, stride=2)   #=> 40x40
        self.conv2 = nn.Conv2d(4, 8, kernel_size=2, stride=2)   #=> 20x20
        self.conv3 = nn.Conv2d(8, 16, kernel_size=2, stride=2)  #=> 10x10

        self.size=16*10*10
        
        # 1 fully connected layer
        self.fc1 = nn.Linear(self.size, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 1)
        self.sig = nn.Sigmoid()
        
    def forward(self, x):
        
    ########
    ## 
    ## Modify your neural network
    ##
    ########
        
        
        x = F.selu(self.conv1(x))
        x = F.selu(self.conv2(x))
        x = F.selu(self.conv3(x))


        x = x.view(-1, self.size)
        x = F.selu(self.fc1(x))  
        x = F.selu(self.fc2(x))  

        x = self.sig(self.fc3(x))
    
        return x

# use your own policy!
policy=Policy().to(device)

#policy=pong_utils.Policy().to(device)

# we use the adam optimizer with learning rate 2e-4
# optim.SGD is also possible
import torch.optim as optim
optimizer = optim.Adam(policy.parameters(), lr=1e-4)
````

## Exercise 2: write your own function for training

### PPO
Later on, you'll implement the PPO algorithm as well, and the scalar function is given by
$\frac{1}{T}\sum^T_t \min\left\{R_{t}^{\rm future}\frac{\pi_{\theta'}(a_t|s_t)}{\pi_{\theta}(a_t|s_t)},R_{t}^{\rm future}{\rm clip}_{\epsilon}\!\left(\frac{\pi_{\theta'}(a_t|s_t)}{\pi_{\theta}(a_t|s_t)}\right)\right\}$

the ${\rm clip}_\epsilon$ function is implemented in pytorch as ```torch.clamp(ratio, 1-epsilon, 1+epsilon)```

````python
def clipped_surrogate(policy, old_probs, states, actions, rewards,
                      discount = 0.995, epsilon=0.1, beta=0.01):

    ########
    ## 
    ## WRITE YOUR OWN CODE HERE
    ##
    ########
    
    steps = len(rewards)
    discount = discount**np.arange(steps)
    # discounting
    discounted_rewards = np.asarray(rewards)*discount[:,np.newaxis]

    # convert rewards to future rewards, they are the accumulated sum of future discounted rewards.
    rewards_future = discounted_rewards[::-1].cumsum(axis=0)[::-1]
    
    # Mean and std of rewards for normalization
    mean = np.mean(rewards_future, axis=1)
    std = np.std(rewards_future, axis=1) + 1.0e-10
    
    # subtract mean, divide by std dev.
    rewards_normalized = (rewards_future - mean[:,np.newaxis])/std[:,np.newaxis]
    
    # convert everything into pytorch tensors and move to gpu if available
    actions = torch.tensor(actions, dtype=torch.int8, device=device) # 320x8, 4or5 note:'RIGHTFIRE' = 4 and 'LEFTFIRE" = 5 
    old_probs = torch.tensor(old_probs, dtype=torch.float, device=device) #320x8 -> P(right)
    rewards = torch.tensor(rewards_normalized, dtype=torch.float, device=device) #320x8

    # convert states to policy (or probability)
    # Note, the states are a tensor with the size of 320x8x2x80x80 where 320 is the number of steps
    # from the previous cell, 8 is the number of agents, 2 is the number of channels (black and white),
    # and 80x80 are the dimensions of the input pictures.
    states = torch.stack(states)

    # reshaping the states to use them as inputs to 2560x2x80x80 note: 2560=320x8
    policy_input = states.view(-1,*states.shape[-3:])
   
    #policy input -> output from nn -> reshape to 320x8 (probs of 'RIGHTFIRE')
    new_probs = policy(policy_input).view(states.shape[:-3])
 
    # update step with new probabilities -> where the action is 4 -> p, otherwise (1-p)
    new_probs = torch.where(actions == 4, new_probs, 1.0-new_probs)
 
    ratio = new_probs/old_probs
    
    # clipping
    clip = torch.clamp(ratio, 1-epsilon, 1+epsilon)
    clipped = torch.min(ratio*rewards, clip*rewards)
    


    # include a regularization term
    # this steers new_policy towards 0.5
    # prevents policy to become exactly 0 or 1 helps exploration
    # add in 1.e-10 to avoid log(0) which gives nan
    entropy = -(new_probs*torch.log(old_probs+1.e-10)+ \
        (1.0-new_probs)*torch.log(1.0-old_probs+1.e-10))

    return torch.mean(clipped + beta*entropy)
````

# Training
We are now ready to train our policy!
WARNING: make sure to turn on GPU, which also enables multicore processing. It may take up to 45 minutes even with GPU enabled, otherwise it will take much longer!

````python
from parallelEnv import parallelEnv
import numpy as np
# keep track of how long training takes
# WARNING: running through all 800 episodes will take 30-45 minutes

# training loop max iterations
episode = 3500

# widget bar to display progress
#!pip install progressbar
import progressbar as pb
widget = ['training loop: ', pb.Percentage(), ' ', 
          pb.Bar(), ' ', pb.ETA() ]
timer = pb.ProgressBar(widgets=widget, maxval=episode).start()


envs = parallelEnv('PongDeterministic-v4', n=8, seed=1234)

discount_rate = .99
epsilon = 0.1
beta = .01
tmax = 320
SGD_epoch = 4

# keep track of progress
mean_rewards = []

for e in range(episode):

    # collect trajectories
    old_probs, states, actions, rewards = \
        pong_utils.collect_trajectories(envs, policy, tmax=tmax)
        
    total_rewards = np.sum(rewards, axis=0)


    # gradient ascent step
    for _ in range(SGD_epoch):
        
        # uncomment to utilize your own clipped function!
        L = -clipped_surrogate(policy, old_probs, states, actions, rewards, epsilon=epsilon, beta=beta)

        #L = -pong_utils.clipped_surrogate(policy, old_probs, states, actions, rewards,
        #                                 epsilon=epsilon, beta=beta)
        
        optimizer.zero_grad()
        L.backward()
        optimizer.step()
        del L
    
    # the clipping parameter reduces as time goes on
    epsilon*=.999
    
    # the regulation term also reduces
    # this reduces exploration in later runs
    beta*=.995
    
    # get the average reward of the parallel environments
    mean_rewards.append(np.mean(total_rewards))
    
    # display some progress every 150 iterations
    if (e+1)%150 ==0 :
        print("Episode: {0:d}, score: {1:f}".format(e+1,np.mean(total_rewards)))
        print(total_rewards)
        
    # update progress widget bar
    timer.update(e+1)
    
timer.finish()
````
